#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

#### Ans:-Flask is a lightweight and flexible web application framework written in Python. It is classified as a microframework because it does not require particular tools or libraries and allows developers to create web applications quickly and easily. Flask is suitable for small to medium-sized projects, prototypes, and smaller applications.

Advantages of Flask Framework:

#### Lightweight: Flask is a lightweight framework and has a small footprint. It does not require particular tools or libraries and allows developers to create web applications quickly and easily.

#### Easy to Learn: Flask has a simple and easy-to-learn syntax. It does not require complex configurations or setup, making it an ideal choice for beginners.

#### Flexible: Flask is highly flexible and allows developers to build applications the way they want. It offers a variety of extensions and libraries that can be used to add additional functionality to the application.

#### Modular: Flask is a modular framework and allows developers to use only what they need. It offers a variety of plugins that can be used to add additional functionality to the application.

#### Scalable: Flask is highly scalable and can handle small to medium-sized projects with ease. It can be used to build complex applications that can be easily scaled up or down based on demand.

##### Extensible: Flask offers a variety of extensions and libraries that can be used to add additional functionality to the application. It supports third-party libraries and can be easily integrated with other technologies.

#### Large Community: Flask has a large and active community of developers who contribute to its development and provide support to other developers. This makes it easy to find resources and solutions to common problems.

#### Q3. What is App routing in Flask? Why do we use app routes?

#### Ans:-App routing is a feature of the Flask web framework that allows developers to map URLs to Python functions, also known as view functions. In other words, app routing defines the behavior of the application when a particular URL is requested.

@app.route('/hello')
def hello():
    return 'Hello, World!'


Here are some benefits of using app routes in Flask:

#### URL organization: App routing allows developers to organize the application's URLs and view functions in a structured manner. This makes it easier to maintain and update the application.

#### Reusability: By defining app routes, developers can reuse the same code across different URLs, which saves time and effort.

#### Easy testing: App routing makes it easier to test the application's functionality by isolating different URLs and view functions.

##### Clear code structure: App routing helps developers to maintain a clear and organized code structure, which improves readability and makes debugging easier.

 app routing is an essential feature of Flask that helps developers to create well-structured and organized web applications that can handle different requests and responses.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

#### Ans:-In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint with any parameters encoded as URL arguments. The url_for() function takes the name of the endpoint as the first argument and any additional arguments representing the URL parameters.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

@app.route('/user/<username>')
def profile(username):
    return f'{username}\'s profile'

@app.route('/url')
def url_demo():
    url1 = url_for('hello', name='John')
    url2 = url_for('profile', username='jdoe')
    return f'URL for hello(): {url1} <br> URL for profile(): {url2}'

if __name__ == '__main__':
    app.run()
```
