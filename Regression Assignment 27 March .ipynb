#### Q1. Explain the concept of R-squared in linear regression models. How is it calculated, and what does it represent?

#### Ans:-R-squared (R2) is a statistical measure that represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model.


### R² = 1 - (RSS / TSS)
The R-squared value ranges from 0 to 1, where 0 indicates that the model does not explain any of the variation in the dependent variable, and 1 indicates that the model explains all of the variation in the dependent variable

- R-squared (R²) is a measure of how well a linear regression model fits the data. It ranges from 0 to 1, where 0 means the model doesn't explain any of the variation in the data, and 1 means the model perfectly explains all the variation.

#### Q2. Define adjusted R-squared and explain how it differs from the regular R-squared.

#### Ans:-Adjusted R-squared is a modified version of R-squared that has been adjusted for the number of predictors in the model. The adjusted R-squared increases when the new term improves the model more than would be expected by chance. It decreases when a predictor improves the model by less than expected.

## Adjusted R² = 1 - [(1-R²)(n-1)/(n-p-1)]

##### _The adjusted R-squared penalizes the inclusion of unnecessary predictor variables in the model and thus helps to avoid overfitting. It is always lower than the regular R-squared and decreases when additional variables are added to the model that do not improve the model's fit.

#### Q3. When is it more appropriate to use adjusted R-squared?

#### Ans:-Adjusted R-squared is more appropriate to use when the number of predictor variables in a linear regression model is large or when the sample size is small. In these situations, the regular R-squared tends to overestimate the model's goodness of fit and may give a misleading indication of the model's predictive power.

For example, if a model has many predictor variables and a high regular R-squared value, it may be overfitting the data and not generalizing well to new data. In contrast, a model with a lower adjusted R-squared value indicates that some of the predictor variables are not contributing significantly to the model's fit, and removing them could improve the model's predictive power.

#### Q4. What are RMSE, MSE, and MAE in the context of regression analysis? How are these metrics calculated, and what do they represent?

##### Ans:- Mean Squared Error (MSE): MSE measures the average squared difference between the predicted and actual values. It is calculated by taking the sum of the squared differences between the predicted and actual values, and then dividing by the number of observations.

### MSE = (1/n) * Σ(yi - ŷi)²
-  n = is the number of observations,
-  yi= is the actual value of the dependent variable, 
- ŷi= is the predicted value of the dependent variable.

#### - Root Mean Squared Error (RMSE): RMSE is the square root of the MSE and is used to measure the average distance between the predicted and actual values. It is calculated by taking the square root of the MSE.

### RMSE = √(MSE)

#### - Mean Absolute Error (MAE): MAE measures the average absolute difference between the predicted and actual values. It is calculated by taking the sum of the absolute differences between the predicted and actual values, and then dividing by the number of observations.

### MAE = (1/n) * Σ|yi - ŷi|

- a regression model is performing by comparing the predicted values to the actual values. The lower the value of RMSE, MSE, or MAE, the better the model is performing in terms of accuracy.

#### Q6. Explain the concept of Lasso regularization. How does it differ from Ridge regularization, and when is it more appropriate to use?

#### Ans:-Lasso regularization is a method used to prevent overfitting in linear regression models by adding a penalty term to the loss function. The penalty term is based on the absolute values of the model's coefficients, and it forces some of the coefficients to be exactly zero. This has the effect of shrinking some coefficients towards zero, effectively removing them from the model and reducing the model's complexity. 


- Lasso regularization differs from Ridge regularization in that the penalty term is based on the absolute values of the coefficients, rather than their squares. As a result, Lasso tends to produce sparser models than Ridge, with more coefficients exactly equal to zero.

- Lasso regularization is more appropriate to use when there are many irrelevant or redundant features in the data that can be safely removed from the model without sacrificing too much predictive power. It is also useful when interpreting the model's coefficients is important, as the sparsity of the model makes it easier to identify the most important features.

#### Q8. Discuss the limitations of regularized linear models and explain why they may not always be the best choice for regression analysis.

#### Ans:-Regularized linear models such as Ridge and Lasso regression can be powerful tools for regression analysis, but they do have limitations and may not always be the best choice for every situation.

#### Loss of Interpretability: Regularization involves adding a penalty term to the objective function that the model is trying to optimize. This penalty term can shrink the coefficients of some features towards zero, effectively removing them from the model.

#### Limited Flexibility: Regularization assumes a linear relationship between the input features and the target variable

#### Hyperparameter Tuning: Regularization involves tuning a hyperparameter that controls the strength of the penalty term. 

#### Limited Effectiveness for High-Dimensional Data: Regularization works best when the number of features is less than the number of observations.

#### Biased Predictions: Regularization can bias the model towards certain types of predictions, such as predictions that are closer to zero or predictions that have smaller magnitudes.
