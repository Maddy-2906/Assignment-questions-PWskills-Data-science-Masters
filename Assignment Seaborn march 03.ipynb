#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Ans:-Scatter plot: A scatter plot is used to visualize the relationship between two variables. It is often used to identify patterns in the data or to look for trends.

#### Line plot: A line plot is a basic plot type that displays data points as markers connected by straight lines. It is used to visualize the trend of a variable over time or to show the relationship between two variables.

#### Histogram: A histogram is used to visualize the distribution of a single variable. It is a plot that consists of bars, where the height of each bar represents the number of data points that fall within a particular range.

#### Box plot: A box plot is used to display the distribution of a variable. It displays the median, quartiles, and outliers of a variable in a concise manner.

#### Heatmap: A heatmap is a plot that is used to visualize the relationship between two variables in a two-dimensional plot. It is often used to identify the patterns in the data or to visualize the correlations between variables.

```python
flights = sns.load_dataset('flights')
```

```python
import seaborn as sns

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Reshape the dataset to create a pivot table
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")
```
