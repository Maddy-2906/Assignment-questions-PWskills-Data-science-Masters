#### Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?

#### Ans:-There are primarily five assumptions of linear regression. 

- There is a linear relationship between the predictors (x) and the outcome (y)
- Predictors (x) are independent and observed with negligible error
- Residual Errors have a mean value of zero
- Residual Errors have constant variance
- Residual Errors are independent from each other and predictors (x)

##### assumptions of linear regression diagnostic checks
- Linear regression assumes a linear relationship between the dependent variable and the independent variable(s).
- The observations should be independent of each other and there should be no pattern in the residuals over time or space.
- Homoscedasticity is assumed, meaning that the variance of the errors should be constant across all values of the independent variable(s).
- Normality of the errors is also assumed, meaning that the distribution of the residuals should be symmetric and bell-shaped.
- multicollinearity, which is high correlation between the independent variables, should be avoided to ensure reliable and stable estimates of the regression coefficients.

#### Q4. Explain the concept of gradient descent. How is it used in machine learning?

#### Ans:-Gradient descent is an optimization algorithm which is commonly-used to train machine learning models and neural networks.  Training data helps these models learn over time, and the cost function within gradient descent specifically acts as a barometer, gauging its accuracy with each iteration of parameter updates. 

- 1.Batch gradient descent computes the gradient of the cost function with respect to all training examples in the dataset at once. This can be computationally expensive for large datasets, but it tends to converge to the global minimum of the cost function.

- 2.Stochastic gradient descent (SGD) computes the gradient of the cost function with respect to one training example at a time. This is faster and can be used for online learning, but it may not converge to the global minimum and can be more sensitive to the learning rate.
The steps of the gradient descent algorithm are as follows:

- Initialize the model parameters to some arbitrary values.
- Calculate the gradient (i.e., the derivative) of the cost function with respect to each parameter.
- Update the model parameters by subtracting a fraction of the gradient from the current parameter values. The fraction is determined by a learning rate hyperparameter, which controls the step size of each update.
- Repeat steps 2-3 until convergence (i.e., until the cost function reaches a minimum or until a maximum number of iterations is reached).

#### In machine learning, gradient descent is used to optimize the parameters of a wide range of models, including linear regression, logistic regression, neural networks, and support vector machines. By minimizing the cost function using gradient descent, we can improve the accuracy and performance of these models on new, unseen data.

#### Q5. Describe the multiple linear regression model. How does it differ from simple linear regression?

#### Ans:Multiple linear regression is a method we can use to quantify the relationship between two or more predictor variables and a response variable.


- Compared to simple linear regression, multiple linear regression allows for more complex relationships between the dependent variable and multiple independent variables. It also allows for the identification of the unique contribution of each independent variable to the dependent variable, while controlling for the effects of other independent variables

- multiple linear regression requires a larger sample size to estimate the coefficients accurately, and it may suffer from multicollinearity, which is the presence of high correlations between the independent variables. To address this issue, techniques such as variable selection, regularization, and principal component analysis can be used.


#### Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?

#### Ans:-Multicollinearity occurs when two or more independent variables in a data frame have a high correlation with one another in a regression model. This means that one independent variable can be predicted from another in a regression model
There are several ways to detect multicollinearity in a multiple linear regression model:

#### Correlation matrix: Calculate the correlation matrix between all pairs of independent variables. If any two variables have a correlation coefficient close to 1 or -1, it indicates that they are highly correlated.

##### Variance Inflation Factor (VIF): Calculate the VIF for each independent variable. VIF measures the extent to which the variance of the coefficient estimate is increased due to multicollinearity. If the VIF for a variable is greater than 5, it suggests that there is significant multicollinearity

To address multicollinearity in a multiple linear regression model, here are some common techniques:
#### Feature selection: Remove one or more highly correlated variables from the model. This can help to reduce the multicollinearity and simplify the model.

#### Principal Component Analysis (PCA): Use PCA to create a smaller set of uncorrelated variables (known as principal components) that capture most of the variation in the original variables. Then use these principal components as the independent variables in the regression model.

#### Ridge regression: Use ridge regression, which is a type of regularization technique that adds a penalty term to the cost function to shrink the coefficients towards zero. This can help to stabilize the estimates of the coefficients and reduce the impact of multicollinearity.

#### VIF-based elimination: Eliminate the variable with the highest VIF and refit the model. Repeat until all VIF values are less than the threshold value (usually 5).

#### Q8. What are the advantages and disadvantages of polynomial regression compared to linear regression? In what situations would you prefer to use polynomial regression?

#### Ans:-Advantages of polynomial regression over linear regression:

- Can model non-linear relationships between the dependent and independent variables.
- Can capture more complex patterns in the data.
- Can provide more accurate predictions when the relationship between the variables is non-linear.
#### Disadvantages of polynomial regression compared to linear regression:
- Can suffer from overfitting if the degree of the polynomial is too high.
- Can be computationally expensive for high-degree polynomials.
- Can be difficult to interpret the coefficients of higher-degree terms.

- polynomial regression is preferred over linear regression when the relationship between the dependent and independent variables is non-linear. It is also useful when higher-order terms are needed to capture the relationship
