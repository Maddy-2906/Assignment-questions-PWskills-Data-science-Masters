#### Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its application.

#### Ans:-Min-Max scaling, also known as normalization, is a common technique used in data preprocessing to transform the numerical features of a dataset so that they are within a specific range. The scaling is performed by subtracting the minimum value of the feature and then dividing by the range of the feature. This results in all the values of the feature being transformed to a range between 0 and 1.

The formula for Min-Max scaling is:

#### scaled_value = (value - min) / (max - min)

```python

import numpy as np

# create dataset
age = np.array([25, 30, 35, 40, 45, 50])
income = np.array([50000, 70000, 45000, 80000, 55000, 90000])

# combine features into a single array
X = np.vstack((age, income)).T

from sklearn.preprocessing import MinMaxScaler

# create MinMaxScaler object
scaler = MinMaxScaler()

# fit scaler to data
scaler.fit(X)

# transform data
X_scaled = scaler.transform(X)

# print scaled data
print(X_scaled)
```

#### Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.

#### Ans:-Principal Component Analysis (PCA) is a dimensionality reduction technique that is commonly used to reduce the number of features or variables in a dataset while preserving the most important information.

#### PCA works by finding the principal components of a dataset, which are linear combinations of the original features that capture the most variation in the data. The first principal component captures the most variation in the data, followed by the second principal component, and so on.

#### PCA can be used in data preprocessing to reduce the number of features and remove redundant information. By reducing the number of features, it can simplify the analysis and reduce the risk of overfitting.

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the iris dataset
iris = load_iris()

# Create a PCA object with two principal components
pca = PCA(n_components=2)

# Apply PCA to the iris dataset
iris_pca = pca.fit_transform(iris.data)

# Print the shape of the original and transformed datasets
print("Original dataset shape:", iris.data.shape)
print("Transformed dataset shape:", iris_pca.shape)
```

#### Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature Extraction? Provide an example to illustrate this concept.

#### Ans:-PCA (Principal Component Analysis) and Feature Extraction are closely related concepts. Feature Extraction is a process of extracting relevant features from the original set of features to improve the performance of the machine learning model. PCA is a popular technique used for feature extraction, and it involves transforming the original features into a new set of features that are linearly uncorrelated, known as principal components.

#### Standardize the data: It is essential to standardize the data before performing PCA to ensure that each feature has a mean of 0 and a variance of 1.

#### Calculate the covariance matrix: The covariance matrix is calculated for the standardized data to determine the relationship between the different features.

#### Calculate the eigenvectors and eigenvalues of the covariance matrix: Eigenvectors and eigenvalues are calculated from the covariance matrix, and they represent the direction and magnitude of the new principal components, respectively.

#### Select the top k principal components: The top k principal components are selected based on their corresponding eigenvalues, with the highest eigenvalues indicating the most significant components.

#### Transform the original features: The original features are transformed into the new set of features represented by the selected principal components.

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# load iris dataset
iris = load_iris()

# standardize the data
scaler = StandardScaler()
iris_std = scaler.fit_transform(iris.data)

# perform PCA with 2 components
pca = PCA(n_components=2)
iris_pca = pca.fit_transform(iris_std)

# print the explained variance ratio of the 2 principal components
print(pca.explained_variance_ratio_)

# plot the transformed data
import matplotlib.pyplot as plt
plt.scatter(iris_pca[:,0], iris_pca[:,1], c=iris.target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

#### X_scaled = (X - X_min) / (X_max - X_min) * (max_range - min_range) + min_range

X is the original value
X_min is the minimum value of the dataset (in this case, 1)
X_max is the maximum value of the dataset (in this case, 20)
max_range is the maximum range of the scaled values (in this case, 1)
min_range is the minimum range of the scaled values (in this case, -1)
Using the above formula, we can scale the values to a range of -1 to 1 as follows:

#### X_scaled = (X - 1) / (20 - 1) * (1 - (-1)) + (-1)
