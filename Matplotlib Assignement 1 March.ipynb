#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

#### Ans:-Matplotlib is a Python library for creating static, animated, and interactive visualizations in Python. It is one of the most widely used libraries for data visualization in Python.

#### Matplotlib is used for creating high-quality plots, graphs, charts, and other visualizations of data. It can be used for data exploration, analysis, and communication of findings. Matplotlib provides a wide range of customization options for creating a wide range of visualizations to suit various needs.

five types of plots that can be plotted using the Pyplot module of Matplotlib:

#### Line plots: A line plot is a simple plot that shows the relationship between two variables as a series of data points connected by straight lines.

#### Scatter plots: A scatter plot is a plot that shows the relationship between two variables as a set of points on a two-dimensional plane.

#### Histograms: A histogram is a plot that shows the distribution of a variable by dividing the data into a set of bins and counting the number of observations in each bin.

#### Bar plots: A bar plot is a plot that shows the relationship between a categorical variable and a numerical variable as a set of bars.

#### Pie charts: A pie chart is a plot that shows the proportion of different categories in a data set as a set of slices in a circular chart.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data

#### ans:-A scatter plot is a type of plot that displays the relationship between two continuous variables. It is a two-dimensional plot that uses dots to represent the data points, with the position of each dot representing the values of the two variables for a particular observation.
In a scatter plot, one variable is plotted on the horizontal axis (x-axis) and the other variable is plotted on the vertical axis (y-axis). Each dot on the plot represents an observation in the data set. The position of the dot on the plot indicates the values of the two variables for that observation.

Scatter plots are useful for exploring the relationship between two variables, such as examining the correlation between them. Scatter plots can also be used to identify any patterns or trends in the data, such as clustering or outliers.

Additionally, scatter plots can be enhanced by adding additional visual elements such as regression lines, markers, colors, or labels to help better understand the relationship between the two variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y,c='#7949EA',alpha=.9)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Dot Graph")
plt.grid()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:

#### Ans:-The subplot() function in Matplotlib is used to create multiple plots in a single figure. It is a useful tool for visualizing and comparing multiple datasets or variables in a single plot.
The subplot() function takes three arguments - the number of rows, the number of columns, and the index of the current plot. The index starts at 1 and increases row-wise, then column-wise. For example, if we want to create a 2x2 grid of plots, we would use subplot(2, 2, 1) for the first plot, subplot(2, 2, 2) for the second plot, and so on.

Using subplot(), we can create a grid of subplots and plot different data sets or variables in each subplot. This allows us to easily compare the different plots and identify any patterns or trends in the data.

Furthermore, the subplot() function provides additional customization options such as adjusting the spacing between the subplots, adding titles or labels to the subplots, and changing the size of the overall figure.

Overall, the subplot() function is a useful tool for creating complex plots with multiple subplots in a single figure, which can be helpful for visualizing and analyzing complex data sets.








```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1 in the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2 in the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3 in the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4 in the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### Ans:-A bar plot is a type of plot in which rectangular bars are used to represent the values of a categorical variable. The height or length of each bar is proportional to the value of the variable it represents. Bar plots can be used to visualize the distribution of a single categorical variable, or to compare the values of a categorical variable across different groups.


A horizontal bar plot is a type of bar plot in which the bars are arranged horizontally instead of vertically.


```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.bar(company, profit)
plt.title("Company profits")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title("Company profits")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

#### Ans:-A box plot is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, maximum, median, and the first and third quartiles. It is also known as a box-and-whisker plot.

Box plots are used to display the distribution of a dataset, including its outliers, skewness, and spread. They provide a quick visual summary of the data and are useful for comparing multiple datasets.


```python
import numpy as np
import matplotlib.pyplot as plt
#data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
data = [box1, box2]
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Data")
plt.ylabel("Value")
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
