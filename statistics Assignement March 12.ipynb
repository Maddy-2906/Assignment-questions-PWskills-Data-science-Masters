#### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

n = 30 # sample size
mean = 50
std = 5
alpha = 0.05

t_value = stats.t.ppf(1 - alpha/2, n - 1)
margin_of_error = t_value * std / (n**0.5)

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Lower bound: {lower_bound:.2f}")
print(f"Upper bound: {upper_bound:.2f}")
```

#### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import scipy.stats as stats
import numpy as np

observed = np.array([50, 45, 55, 25, 25, 50])  # observed frequencies of each color
expected = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed)  # expected frequencies of each color

chi2, p = stats.chisquare(observed, expected)

alpha = 0.05

print(f"Chi-square statistic: {chi2:0.2f}")
print(f"P-value: {p:0.2f}")

if p < alpha:
    print("The distribution of colors in the M&M bag does not match the expected distribution.")
else:
    print("The distribution of colors in the M&M bag matches the expected distribution.")

```

#### Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import statsmodels.api as sm
import scipy.stats as stats
import numpy as np

n = 500  # sample size
smokers = 60  # number of smokers in the sample

prop = smokers / n  # sample proportion
se = np.sqrt(prop * (1 - prop) / n)  # standard error

alpha = 0.05  # significance level
z = stats.norm.ppf(1 - alpha / 2)  # z-score for the confidence level

ci_low, ci_high = prop - z * se, prop + z * se  # confidence interval

print(f"95% Confidence Interval: ({ci_low:0.3f}, {ci_high:0.3f})")
```

#### Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

#### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df=10 #degree of freedom
x=np.linspace(0,30,500)
y=stats.chi2.pdf(x,df)
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x,y,'-b',linewidth=2)
ax.fill_between(x[x>=15],y[x>=15],0,color='green',alpha=0.5)
ax.set_xlabel('Chi-Square Statistic') # x-axis data name
ax.set_ylabel('Probability Density') # y-axis data name
ax.set_title(f"Chi-Square Distributation with {df} Degreebof Freedom")

plt.show()
```

#### Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

#### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

#### Ans:- Null hypothesis: The population mean is equal to 70 ( = 70).
#### Alternative hypothesis: The population mean is significantly different from 70 (  70).

#### We will use a t-test since the population standard deviation is unknown and the sample size is less than 30. The t-test statistic is calculated as:
#### t = (sample mean - hypothesized mean) / (sample standard error)
#### t = (72 - 70) / (10 / sqrt(30)) = 1.095
#### Using a t-distribution table with 29 degrees of freedom and a significance level of 0.05 (two-tailed), the critical values are 2.045.

#### the calculated t-value of 1.095 falls within the range of -2.045 to 2.045, we fail to reject the null hypothesis.

#### Conclusion :-  we do not have sufficient evidence to conclude that the population mean is significantly different from 70 at a significance level of 0.05.
