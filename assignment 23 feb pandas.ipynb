#### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

#### Ans:-In Pandas, a DataFrame is a 2-dimensional labeled data structure with columns of potentially different data types. It is similar to a table in a relational database or a spreadsheet. Each column of a DataFrame is a Pandas Series.

#### A Series in Pandas is a one-dimensional labeled array that can hold data of any type (integer, string, float, etc.) and has an axis label, called the "index". It can be thought of as a single column of data within a DataFrame.

```python

import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30,27],
    'Gender': ['Female', 'Male', 'Female'],
    
}
df = pd.DataFrame(data)
ages = df['Age']

print(df)
print(ages)
```

#### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

#### Ans:-Pandas that can be used to manipulate data in a DataFrame. Here are some commonly used functions:

#### head() and tail(): These functions can be used to display the first few or last few rows of a DataFrame, respectively. For example, you might use df.head() to quickly check the first few rows of a large DataFrame.

#### describe(): This function provides summary statistics of the DataFrame, such as count, mean, standard deviation, minimum, and maximum. You might use df.describe() to get a quick overview of the range of values in each column of a DataFrame.

#### loc[] and iloc[]: These functions can be used to access rows and columns of a DataFrame by label or integer location, respectively. For example, you might use df.loc[3, 'Age'] to access the age value in the row with index label 3 and column label 'Age'.

#### groupby(): This function can be used to group rows of a DataFrame by one or more columns and apply a function to each group. For example, you might use df.groupby('Country')['Age'].mean() to calculate the mean age for each country in the DataFrame.

#### drop(): This function can be used to drop rows or columns from a DataFrame. For example, you might use df.drop('Country', axis=1) to drop the 'Country' column from the DataFrame.

#### fillna(): This function can be used to fill missing values in a DataFrame with a specified value or method. For example, you might use df.fillna(method='ffill') to fill missing values with the last observed value in the DataFrame.

#### Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

#### Ans:-In Pandas, Series and DataFrame are mutable in nature, while Panel is not.

#### A Series can be modified by assigning a new value to a specific index. Similarly, a DataFrame can be modified by assigning new values to specific rows, columns, or individual cells.

#### However, a Panel is a 3-dimensional data structure that is not commonly used in Pandas. It is generally recommended to use DataFrame instead, as it can handle most of the use cases for multi-dimensional data. Panel is not as widely used and is not considered as mutable as a Series or DataFrame.
